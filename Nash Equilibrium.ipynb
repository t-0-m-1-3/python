# We're studying a basic problem of choices
---
simplified, two actors have a choice of 4 options. Each option has a different utility to the actor. 

```python
steve = [0,2,4,1]
viki = [0,0,2,5]
```

```python
a = np.array([steve, viki]).T
a
```

## Above you can see not every payoff is optimal for each actor.
### Someone will be unhappy with a good chance of a 0 sum outcome


```python
fig1 = plt.figure(figsize = (8,5))

ax1 = fig1.add_subplot(111)
ax1.add_patch(patches.Polygon(a, fill = False))
plt.plot(steve, viki, 'o')
plt.xlim(min(steve) -1, max(steve)+1)
plt.ylim(min(viki)-1, max(steve)+1)
plt.axhline(color='black')
plt.axvline(color='black')
plt.title('Payoff Polygon regarrding Chicken')
```

## We need to find the slope of the line that maximizes both points
---
In Nash I trust

Equation Lines
---
To maximize the `xy`, we need an expression for `y` in terms of `x` which can join points on the slope. 

```python
def n(x):
    return -x**2 + 6*x
```

```python
x = sy.Symbol('x')
```

```python
dn = sy.diff(n(x), x)
```

```python
soln = sy.solve(dn)
```

```python
print("The Derivative of the product is ", dn, "\n The Critical Point for x = ", soln, "\nthis corresponds with y = ", -1*(soln[0] - 4)+2)
```
